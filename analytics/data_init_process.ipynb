```python
## Add date column
def ts_str_to_date(ts):
    ts = str(ts)
    mon = ts[4:7]
    day = ts[8:10]
    year = ts[26:30]
    return  mon + '-' + day + '-' + year
    
```

```python
dates_column = [ts_str_to_date(ts) for ts in df_tw.ts.values]
id_int_column = [int(id) for id in df_tw.id_str.values]
df_tw['date'] = dates_column
df_tw['id_int'] = id_int_column
```
